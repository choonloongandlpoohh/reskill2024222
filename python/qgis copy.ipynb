# Xml-python을 이용한 방법

```python
import requests
import pandas as pd
from urllib.parse import urlencode, quote_plus, unquote
import xml.etree.ElementTree as ET
```

```python
url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
key = unquote('%2BWA%2FmwYw0b3ZhWTQXpswD3cuP35%2Ba6yCaQd5nJVoHuqiOYwT%2FeD9ZWfCfsHVKkX4MVHCir32Ry40fJPeDW%2FOBQ%3D%3D')
```

```python
weather3 =[] # 전체 자료를 담을 리스트

#함수정의. stationID, 시작일, 종료일을 받아 weather 리스트에 결과 추가
def query(stnId, startDt, endDt):
    queryParams = '?' + urlencode(
    # urlencode() : url에 문자를 표현하는 문자 인코딩 방법
    # quote_plus() : whitespace를 %20이 아닌 +로 표현
    {
        quote_plus('serviceKey') : key,
        quote_plus('dataCd') : 'ASOS',
        quote_plus('dateCd') : 'DAY',
        quote_plus('startDt') : startDt,
        quote_plus('endDt') : endDt,
        quote_plus('pageNo') : '1'
    })
    
    # 작성된 요청 메시지를 보내서 웹 페이지 요청
response3 = requests.get(url + queryParams)
xmlstring = response3.content
tree3 = ET.ElementTree(ET.fromstring(xmlstring))
    
    # xml 탐색
root3 = tree3.getroot()
    for item3 in root3.findall('./body/items/item') :
        weather0 = [] # 현재 아이템 한줄 생성
        for child in item3 :
            if child.tag == 'stnId' :
                weather0.append(child.text)
                    
            if child.tag == 'stnNm' :
                weather0.append(child.text)
                
            if child.tag == 'tm' :
                weather0.append(child.text)
                    
            if child.tag == 'avgTa' :
                weather0.append(child.text)
                    
            if child.tag == 'minTa' :
                weather0.append(child.text)
                
            if child.tag == 'maxTa' :
                weather0.append(child.text)
            
        weather3.append(weather0) # 전체 자료에 이 한줄을 추가
        weather0.clear

# 필요한 지점마다 함수 호출
query('108', '20220101', '20220131') # 서울
query('90', '20230101', '20230131') # 속초
query('216', '20230601', '20230630') # 태백

# 열별 이름 지정
headers3 = ['No', 'Gubun', 'Date', 'AvgTemp', 'MinTemp', 'MaxTemp']
           
# 을 통해 결과를 데이터프레임 형식으로 저장
weather_df3 = pd.DataFrame(weather3, columns=headers3)

weather_df3
# 데이터프레임 형식으로 저장된 결과를 스프레드시트 형식으로 저장
# weather_df.to_csv('d:/data/weather3.csv')
```
